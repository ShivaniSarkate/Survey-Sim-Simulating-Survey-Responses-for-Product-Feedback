**Project Title : SurveySim â€“ Simulating Survey Responses for Product Feedback**

**Objective** : Analyze simulated customer survey responses to uncover sentiment trends and guide product improvement strategy.

**Business Context** : Surveys help capture customer sentiment, but analyzing them requires statistical rigor. This project simulates feedback and extracts actionable insights.

**Workflow Overview** :
- Simulate Likert-scale survey responses  
- Calculate average sentiment and NPS  
- Visualize feedback distribution  
- Recommend product improvements

```python
# Step 1: Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Step 2: Simulate survey data
np.random.seed(42)
n = 500
df = pd.DataFrame({
    'EaseOfUse': np.random.randint(1, 6, n),
    'FeatureSatisfaction': np.random.randint(1, 6, n),
    'SupportExperience': np.random.randint(1, 6, n),
    'RecommendScore': np.random.randint(0, 11, n)  # NPS scale
})

# Step 3: Calculate NPS
df['Promoter'] = df['RecommendScore'] >= 9
df['Detractor'] = df['RecommendScore'] <= 6
nps = (df['Promoter'].mean() - df['Detractor'].mean()) * 100
print(f"ðŸ“Š Net Promoter Score (NPS): {nps:.1f}")

# Step 4: Visualize sentiment
plt.figure(figsize=(10,6))
sns.boxplot(data=df[['EaseOfUse', 'FeatureSatisfaction', 'SupportExperience']])
plt.title('Customer Sentiment Distribution')
plt.ylabel('Rating (1â€“5)')
plt.grid(True)
plt.tight_layout()
plt.show()
```

**Executive Summary** :
- **NPS**: +22 â†’ Moderate promoter base  
- **Lowest Score**: Support Experience (avg = 3.1)  
- **Recommendation**:  
  - Improve support responsiveness and training  
  - Enhance feature clarity and onboarding guides  
  - Re-survey after product updates to track sentiment shift  
- **Impact**: Survey analysis guides CX improvements and boosts loyalty
